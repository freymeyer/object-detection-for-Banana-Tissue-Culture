```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
from PIL import Image
```

```python
source = r'C:\Users\Frey\Documents\Capstone\prototype\mga gi cut na\sample_1.jpg'
```

```python
img = cv2.imread(source)
RGB_im = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_im)
```

```python
image_gray = cv2.cvtColor(RGB_im, cv2.COLOR_RGB2GRAY)
plt.imshow(image_gray)
```

```python
edged = cv2.Canny(image_gray, 30, 200) 
plt.imshow(edged)
```

```python
contours, hierarchy = cv2.findContours(edged,  cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE) 
```

```python
images = cv2.drawContours(edged, contours, -1, (0, 255, 0), 3) 
plt.imshow(images)
```
